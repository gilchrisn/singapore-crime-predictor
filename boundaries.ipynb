```python
import geopandas as gpd

# Load the GeoJSON file
npc_boundaries = gpd.read_file('resources\singapore-npc-boundaries\SingaporePoliceForceNPCBoundary.geojson')

# Inspect the data
print(npc_boundaries.head())
```

```python
import matplotlib.pyplot as plt

# Plot the GeoDataFrame
npc_boundaries.plot(figsize=(10, 8), edgecolor='black', alpha=0.5)
plt.title("NPC Boundaries of Singapore")
plt.show()
```
